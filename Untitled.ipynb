```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup
import pandas as pd
import time
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
import os

chromedriver = "D:/Web Scrapping/2/chromedriver.exe"
os.environ["webdriver.chrome.driver"] = chromedriver

# Inicializa el controlador de Chrome


def download_html_ct(pagina):
    driver = webdriver.Chrome(chromedriver)
    driver.get("https://amarillas.emol.com/abogados?p={}".format(pagina))
    time.sleep(5)
    with open(r'pagina{}.html'.format(pagina), 'w', encoding="utf-8") as f:
        f.write(driver.page_source)
    driver.quit()
try: 
    for i in range (3):
        download_html_ct(i)      
except:
    pass
```

# Guardar las url de cada pagina y guardarlas

```python
import glob
import pandas as pd
from bs4 import BeautifulSoup

df = pd.DataFrame(columns=["URL"])

for i in range(0,2):
    url = "D:\Web_Scrapping\Prueba\pagina{}.html".format(i)

    with open(url, encoding="utf-8") as fp:
        soup = BeautifulSoup(fp, 'html.parser')
        section = soup.find("section", {"class": "template-search show-results col_main"})
        urls = section.find_all("a", attrs={"ng-href": True,"ng-click": True,"href": True,"itemprop":True})

        for url in urls:
            df = df.append({"URL": "https://amarillas.emol.com{}".format(url["href"])}, ignore_index=True)

df.to_excel("resultados_url.xlsx", index=False)
```

# Abrir cada url para leer telefonos

```python
def download_html_ct(url,x):
    driver = webdriver.Chrome(chromedriver)
    driver.get(url)
    time.sleep(5)
    with open(r'abogado{}.html'.format(x), 'w',encoding="utf-8") as f:
        f.write(driver.page_source)
    driver.quit()

df = pd.read_excel("resultados_url.xlsx")
urls = df["URL"].tolist()

x=0
try: 
    for url in urls:
        x=x+1
        download_html_ct(url,x)      
except:
    pass
```
